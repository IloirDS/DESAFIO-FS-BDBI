## Plotando um grafico de barras e exibindo a quantidade de funcionÃ¡rios em cada setor

```python
#Plotando um grafico de barras e exibindo a quantidade de funcionÃ¡rios em cada setor:
df['Department'].value_counts().plot(kind='bar', color = 'y')
```

## Plote um grÃ¡fico de linha que exiba a quantidade de funcionÃ¡rios para cada nÃ­vel de escolaridade

```python
#Plotando um grÃ¡fico de linha que exibe a quantidade de funcionÃ¡rios para cada nÃ­vel de escolaridade:
df['Education'].value_counts().sort_index().plot(kind='line', marker = 'o', color = '0')
```

## Plote um grÃ¡fico de pizza  que exiba a quantidade de funcionÃ¡rios que tem ou nÃ£o algum atrito na empresa

```python
#Plotando um grÃ¡fico de pizza que exibe a quantidade de funcionÃ¡rios que tem ou nÃ£o algum atrito na empresa:
df['Attrition'].value_counts().plot(kind='pie')
```
