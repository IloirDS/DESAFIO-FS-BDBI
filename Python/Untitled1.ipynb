## Importando biblioteca Pandas e Matplotlib

```python
#Importando biblioteca Pandas e Matplotlib:
import pandas as pd
import matplotlib.pyplot as plt
```

## Criando DF lendo o arquivo csv

```python
#Criando DF lendo o arquivo csv:
df = pd.read_csv('/content/dados_rh.csv')
df
```

## Vendo quantas linhas e colunas tem o banco de dados

```python
#Vendo quantas linhas e colunas tem o banco de dados:
df.shape
```

## Conferindo se existe dados nulos

```python
#Conferindo se existe dados nulos:
df.isnull().sum()
```

## Conferindo qual o tipo de dado de cada coluna

```python
#Conferindo qual o tipo de dado de cada coluna:
df.dtypes
```

## Plotando um grafico de barras e exibindo a quantidade de funcionários em cada setor

```python
#Plotando um grafico de barras e exibindo a quantidade de funcionários em cada setor:
df['Department'].value_counts().plot(kind='bar', color = 'y')
```

## Plote um gráfico de linha que exiba a quantidade de funcionários para cada nível de escolaridade

```python
#Plotando um gráfico de linha que exibe a quantidade de funcionários para cada nível de escolaridade:
df['Education'].value_counts().sort_index().plot(kind='line', marker = 'o', color = '0')
```

## Plote um gráfico de pizza  que exiba a quantidade de funcionários que tem ou não algum atrito na empresa

```python
#Plotando um gráfico de pizza que exibe a quantidade de funcionários que tem ou não algum atrito na empresa:
df['Attrition'].value_counts().plot(kind='pie')
```
